# U.S. Medical Insurance Costs

```python
import csv 
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

with open("insurance.csv", "r") as insurance_data:
    insurance = csv.DictReader(insurance_data)
    for col in insurance:
        age.append(col["age"])
        sex.append(col["sex"])
        bmi.append(col["bmi"])
        children.append(col["children"])
        smoker.append(col["smoker"])
        region.append(col["region"])
        charges.append(col["charges"])

print("Age: " , age)
print("Sex: ", sex)
print("BMI: ", bmi)
print("Children: ", children)
print("Smoker: ", smoker)
print("Region: ", region)
print("Charges: ", charges)

us_medical_insurance_dictionary = {}

for i in range(len(age)):
    value = us_medical_insurance_dictionary.setdefault(str(i + 1) , {"Age": age[i], "Sex": sex[i], "BMI": bmi[i], "Children": children[i], "Smoker": smoker[i], "Region": region[i], "Charges": charges[i] })

print(us_medical_insurance_dictionary)
max_charge = 0
for charge in charges:
    if float(charge) > float(max_charge):
        max_charge = charge
print("The highest medical insurance cost recorded is :" + str(max_charge))
    
```

```python
#relation between age and cost of insurance 
# plotting graph between age and  charges 
# 1. arrange the age in the ascending order and then the corresponding charges also.
age_new = [int(float(i)) for i in age]
charges_new = [int(float(i)) for i in charges]
age_charges_dict = dict(zip(age_new, charges_new))
import matplotlib.pyplot as plt
age_charges_list = age_charges_dict.items()
age_charges_list = sorted(age_charges_list)
x, y = zip(*age_charges_list)
plt.scatter(x, y)
plt.xlabel("Age")
plt.ylabel("First entry of the Charges for the given age.")
plt.show()
```

```python
age_charges_list_combined = list(zip(age_new, charges_new))
age_charges_list_sorted = sorted(age_charges_list_combined)
totals = {}
for uid, x in age_charges_list_sorted:
    if uid not in totals:
        totals[uid] = x
    else:
        totals[uid] += x

print(totals)
for key, value in totals.items():
    key_str = str(key)
    count = age.count(key_str)
    totals[key] = int(float(value/ count))
print(totals)
import matplotlib.pyplot as plt
totals = totals.items()
x, y = zip(*totals)
plt.scatter(x, y)
plt.xlabel("Age")
plt.ylabel("Aaverage of the total charges for the given age")
plt.show()
```

```python
#to find which region has the highest number of smokers.
#first to find the number of regions listed in the data
smoker_region_list = list(zip(smoker, region))
print(smoker_region_list)
region_dict = {}
for i in range(len(region)):
    key = region_dict.setdefault(region[i])
print(region_dict)
print(region.count("south"))
smoker_region_confirmed = []
for i in range(len(smoker_region_list)):
    if smoker_region_list[i][0] == "yes":
        smoker_region_confirmed.append(smoker_region_list[i])
print(smoker_region_confirmed)

for i in range(len(smoker_region_confirmed)):
    if smoker_region_confirmed[i][1] == "southwest":
        region_dict["southwest"] = smoker_region_confirmed.count(("yes", "southwest"))
    elif smoker_region_confirmed[i][1] == "southeast":
        region_dict["southeast"] = smoker_region_confirmed.count(("yes", "southeast"))
    elif smoker_region_confirmed[i][1] == "northwest":
        region_dict["northwest"] = smoker_region_confirmed.count(("yes", "northwest"))
    elif smoker_region_confirmed[i][1] == "northeast":
        region_dict["northeast"] = smoker_region_confirmed.count(("yes", "northeast"))
print(region_dict)
labels = list(region_dict.keys())
values = list(region_dict.values())

plt.pie(values, labels=labels, autopct='%1.0f%%')
plt.title("Number Of Smokers In Region")
plt.show()
```
